```python
from pathlib import Path
import PyPDF2
import fitz
import numpy as np

def pdf_extract(path_to_files:str):    
    path_to_files = Path(path_to_files)  # convert the str to a pathlib object
    file_nums = 0
    for file in path_to_files.rglob('*.pdf'):  # pathlib has rglob
        pdf = fitz.open(file)
        file_nums +=1
        text = ""
        for page in pdf:
            text += page.getText()
            with open (str(file)[:-4]+'.txt','w',encoding='utf-8') as text_file:
                text_file.write(text)                
    return(file_nums)

path_to_files = r"./pdfs"
# call the function
records = pdf_extract(path_to_files)
print('Total PDFs converted to txt files: ', records)
```

```python
import os
import io
import glob
import numpy as np

path_to_files = r"./pdfs"

chinum = []
diag = []
inter = []
medic = []
lastseen =[]
health = [] 
gender = []

def txt_extract(path_to_files:str):    
    path_to_files = Path(path_to_files)  # convert the str to a pathlib object
    file_nums = 0
    for file in path_to_files.rglob('*.txt'):  # pathlib has rglob
        with io.open(file, encoding="utf-8", errors='replace') as fd:
            content = fd.read()
            content = content.lower()
            content = [content]
            ##extract CHI
            chinum.append(extract_chinumber(content))
            ##extract medication
            medic.append(extract_medication(content))
            ##extract diagnoses
            diag.append(extract_diagnoses(content))
            ##extract intervention
            inter.append(extract_intervention(content))
            ##extract lastseen
            lastseen.append(extract_clinicdate(content))
            ##extract healthboard
            health.append(extract_healthboard(content))
            ##extract gender
            gender.append(extract_gender(content))
        file_nums += 1
    return(file_nums)

records = txt_extract(path_to_files)
print('Total txt files parsed: ', records)
```
